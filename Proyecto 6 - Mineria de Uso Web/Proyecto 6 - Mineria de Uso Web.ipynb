```python
import numpy as np
import pandas as pd
import datetime
```

#### Lectura de csv

```python
data = pd.read_csv('./weblog.csv')
data.head()
```

#### Limpieza de IPS

```python
data = data[data.IP.str.contains('[0-9]*\.')]
data.head()
```

### Limpieza de objetos Date Time

```python
data['Time'] = data['Time'].str.strip("[")
data['Time'] = data['Time'].str.replace(":",' ',1)
data
```

```python
data = data[data.Time.str.contains('[0-9]{2}/[a-zA-Z]{3}/[0-9]{4}')]
data['Time']=pd.to_datetime(data['Time'])
data
```

### Limpieza de URLs
Eliminar obtencion de archivos vizuales

```python
exts = ['gif','jpeg', 'GIF', 'JPEG', 'jpg', 'JPG', 'map','js','css','woff']
for ext in exts:
    data = data[~data.URL.str.contains(ext)]
urls = data.URL.str.split('?',1).to_numpy()
urls
```

#### Obtencion de URLS


```python
urls = [ url[0].replace(' HTTP/1.1','') for url in urls ]
urls[:10]
```

```python
import csv
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
data_or = pd.read_csv('bread basket.csv') 
data_or.head() 
```

```python
data = data_or.drop(columns=['date_time', 'weekday_weekend'])
```

```python
morning = (data[data['period_day'] =="morning"])
morning
```

```python
after = (data[data['period_day'] =="afternoon"]
       .groupby(['Transaction','Item'])
       .count().unstack().reset_index().fillna(0)
       .set_index('Transaction'))
after.columns = after.columns.get_level_values(1)
after 
```

```python
eve = (data[data['period_day'] =="evening"])
eve
```

```python
ngt = (data[data['period_day'] =="night"]
       .groupby(['Transaction','Item'])
       .count().unstack().reset_index().fillna(0)
       .set_index('Transaction')
      )

ngt.columns = ngt.columns.get_level_values(1)
ngt
```

```python
def transformar_uno_cero(x):
    if x <= 0:
        return 0
    else:
        return 1
```

```python
basket_sets = after.applymap(transformar_uno_cero)
frq_items = apriori(basket_sets, min_support = 0.05, use_colnames = True) 
frq_items
```

```python
rules = association_rules(frq_items, metric="lift", min_threshold=.05)
rules.head()
```

```python
def bread_basket(df,support):
    df_local = df.copy()
    df_local = df_local.groupby(['Transaction','Item']).count().unstack().reset_index().fillna(0).set_index('Transaction')
    df_local.columns = df_local.columns.get_level_values(1)
    basket_sets = df_local.applymap(transformar_uno_cero)
    frq_items = apriori(basket_sets, min_support = support, use_colnames = True) 
    rules = association_rules(frq_items, min_threshold=support)
    return frq_items,rules
```

```python
items,reglas = bread_basket(eve,0.05)
items
```

```python
groc = pd.read_csv('groceries - groceries.csv')
groc = groc.drop(columns=['Item(s)'])
ls = groc.fillna('0')
lista = []
for i in ls.values.tolist():
    lista += i
lista.remove('0')
lista = set(lista)
lista.remove('0')
df = pd.DataFrame(columns=lista)
```

```python
for i in ls.values.tolist():
    row = {}
    for item in i:
        if item != '0':
            row[item] = 1
    df = df.append(row, ignore_index=True)
```

```python
df = df.fillna(0)
df
```

```python
frq_items = apriori(df, min_support = 0.05, use_colnames = True) 
rules = association_rules(frq_items, min_threshold=0.05)
rules
```

```python
transactions = []
items = []
with open('groceries.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        transactions.append(row)
        for i in row:
            items.append(i)
        line_count += 1
items = set(items)
```

```python
groc_2 = pd.DataFrame(columns = items)
for tran in transactions:
    row = {}
    for item in tran:
        row[item] = 1
    groc_2 = groc_2.append(row, ignore_index=True)
groc_2 = groc_2.fillna(0)
groc_2
```

```python
frq_items = apriori(groc_2, min_support = 0.05, use_colnames = True) 
rules = association_rules(frq_items, min_threshold=0.05)
rules
```
